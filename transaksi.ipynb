```python
import pandas as pd
import matplotlib.pyplot as plt
import calendar
```

```python
df_t = pd.read_csv("/content/drive/MyDrive/dataset_visualisasidata/transaksi.csv", names=['Date', 'ID', 'Quantity', 'Price'])
```

```python
df_t.Date = pd.to_datetime(df_t.Date)
df_t.year = df_t.Date.dt.year
df_t.month = df_t.Date.dt.month.apply(lambda x: calendar.month_name[x])
data_year = df_t[df_t.year == 2017]
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data_year.month, data_year.Quantity)
plt.title('Transaksi 2017')
plt.xlabel('Bulan')
plt.ylabel('Qty')
plt.xticks(rotation=90)
plt.grid()
plt.show()
```

```python
data_june = data_year [df_t.month == 'June']

plt.figure(figsize=(16, 10))
plt.scatter(data_june.Quantity, data_june.Date)
plt.xlabel('Qty')
plt.ylabel('Tanggal')
plt.title('Transaksi Bulan Juni 2017')
plt.grid()
plt.show()
```
