```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_p = pd.read_csv("/content/drive/MyDrive/dataset_visualisasidata/produk.csv", nrows=20)
```

```python
data = df_p.produk
x = []
y = []
z = []

for i in range(len(data)):
  parts = data[i].split("|")
  x.append(parts[1])
  y.append(parts[2])
  z.append(parts[0])
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y);
plt.title('Data Produk')
plt.xlabel('Nama Produk', fontsize=12)
plt.ylabel('Harga', fontsize=12)
plt.xticks(rotation=80 , fontsize=6)
plt.grid()
plt.show()
```
