```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_kpb = pd.read_csv("/content/drive/MyDrive/dataset_visualisasidata/kebutuhan_per_bulan.csv")
```

```python
df_kpb
```

```python
x = df_kpb.Propinsi
y = df_kpb.tahun
z = df_kpb.kebutuhan
```

```python
plt.figure(figsize=(18, 12))
plt.scatter(x, y, s=10);
for i in range(len(x)):
    plt.text(x[i], y[i], z[i], ha='center', va='bottom', rotation=45)
plt.title('Jumlah kebutuhan per tahun tiap provinsi')
plt.xlabel('Propinsi', fontsize=18)
plt.ylabel('Tahun', fontsize=18)
plt.xticks(rotation=90 , fontsize=6)
plt.grid()
plt.show()
```
