```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_sk = pd.read_csv("/content/drive/MyDrive/dataset_visualisasidata/supply_kantor.csv")
df_sk
```

```python
x = df_sk.Jenis
y = df_sk.Jumlah
z = (df_sk['Harga Satuan'] * df_sk.Jumlah)
```

```python
plt.figure(figsize=(18, 12));
plt.scatter(x, y, s=100);
for i in range(len(x)):
    total = 'Rp.' + str(z[i])
    plt.text(x[i], y[i], total, ha='center', va='bottom', rotation=45)
plt.title('Data Supply Kantor', fontsize=24)
plt.xlabel('Jenis Barang', fontsize=24)
plt.ylabel('Jumlah', fontsize=24)
plt.xticks(rotation=45 , fontsize=18)
plt.grid()
plt.show()
```
