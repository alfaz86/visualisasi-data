```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_pp = pd.read_csv("/content/drive/MyDrive/dataset_visualisasidata/profile_pelanggan.csv")
```

```python
df_pp
```

```python
x = df_pp.provinsi
y = df_pp.jumlah
```

```python
plt.figure(figsize=(18, 14))
plt.scatter(x, y)
plt.title('Data profil pelanggan')
plt.xlabel('Provinsi', fontsize=18)
plt.ylabel('Jumlah', fontsize=18)
plt.xticks(rotation=90 , fontsize=6)
plt.grid()
plt.show()
```
